# **MODELING CLUSTERING**

```python
# euclidean distances function, to find the distance from one point to another point
def euclidean_distances(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))
```

## **KMeans Class**

```python
class KMeans:
    def __init__(self, K, max_iteration):
        # initiations constructor parameter
        self.K = K
        self.max_iteration = max_iteration
        
        # list of sample / data indices for each cluster => for each cluster we asign an empty list
        self.clusters = [[] for _ in range(self.K)]
        # mean feature vector for each cluster => create centroids list
        self.centroids = []
    
    def predict(self, X):
        # X == data to be clustered
        self.X = X
        # number of data, number of features
        self.num_of_data, self.num_of_features = X.shape
        
        # initialize centroids => array size self.K, pick random choice between 0 and the number of data
        # random_data_indexes => choose a random value from 0 to the number of data (jumlah data)
        random_data_indexes = np.random.choice(self.num_of_data, self.K, replace=False)
        # The initial centroid is filled with random data taken from the data variable X
        self.centroids = [self.X[index] for index in random_data_indexes]
        
        # Loop until the max iteration or, until there are no more changes in each centroids
        for _ in range(self.max_iteration):
            # update clusters
            self.clusters = self.create_new_clusters(self.centroids)
            
            # update centroids
            old_centroids = self.centroids
            self.centroids = self.get_cluster_centroids(self.clusters)
            
            # check if no more changes in each centroids
            changes_distances_centroids = 0
            for i in range(self.K):
                changes_distances_centroids += euclidean_distances(old_centroids[i], self.centroids[i])
            if changes_distances_centroids == 0: # if no more changes in each centroids
                break
        
        # CREATE CLUSTERS LABEL (for elbow method)
        # Create new array which has a size of "number of data", without initializing entries.
        cluster_labels = np.empty(self.num_of_data)
        
        for cluster_index, cluster in enumerate(self.clusters): # clusters is list of list (2 dimensional list)
            # cluster is a list (1 dimensional list)
            for data_index in cluster:
                cluster_labels[data_index] = cluster_index
        return cluster_labels
    
        
    def create_new_clusters(self, centroids):
        # initialization clusters list (2 dimensional list) => a list containing K empty lists
        clusters = [[] for _ in range(self.K)]
        
        # enumerate function give us the current index and the current data
        for index, data in enumerate(self.X):
            # find closes centroid
            centroid_index = self.get_closest_cluster_centroid(data, centroids)
            clusters[centroid_index].append(index)
        return clusters
    
    
    def get_closest_cluster_centroid(self, data, centroids):
        min_distance = float('inf') # => minimum distance from one data point to the centroids
        i = 0 # i == index for looping
        closest_centroid_index = i
        for centroid in centroids:
            curr_distance = euclidean_distances(data, centroid) # current distances
            if curr_distance < min_distance:
                min_distance = curr_distance
                closest_centroid_index = i
            i+=1
        return closest_centroid_index
    
    
    def get_cluster_centroids(self, clusters):
        centroids = np.zeros((self.K, self.num_of_features))
        # cluster_index will contain values from 0 to K - 1, which will indicate available clusters
        # cluster => one-dimensional list that indicates a collection of a cluster, which contains the index of X data (meaning the data at this index is part of this cluster)
        for cluster_index, cluster in enumerate(clusters):
            cluster_mean = np.mean(self.X[cluster], axis=0)
            centroids[cluster_index] = cluster_mean
        return centroids
    
    
    def plot(self):
        plt.figure(figsize=(12,10))
        labels= []
        for i in range(self.K):
          labels.append('cluster ' + str(i + 1))
        i = 0
        for cluster in self.clusters:
            # change the shape of array, from an array that has 2 columns (x1 x2) to an array that has 2 rows [[x1 x1 x1 x1 x1 x1 x1 ... x1], [x2 x2 x2 x2 x2 x2 x2 x2 ... x2] ]
            points_of_clusters = self.X[cluster].T
            # points_of_clusters[0] => x axis (feature 1) , points_of_clusters[1] => y axis (feature 2)
            plt.scatter(points_of_clusters[0], points_of_clusters[1],label=labels[i],alpha = 0.8)
            i+=1
        
        # do for loop as many as the number of existing centroids (K centroid)
        for points_of_centroid in self.centroids:
            # points_of_centroid[0] => x axis , points_of_centroid[1] => y axis
            plt.scatter(points_of_centroid[0],points_of_centroid[1], marker="*",s = 300, color="black", linewidth=2)
        plt.title('Dealer customer data cluster')
        plt.xlabel('Lama Berlangganan')
        plt.ylabel('Umur')
        labels.append('Centroids')
        plt.legend(labels, facecolor="white",framealpha=1)
        plt.show()
    
```

## **Elbow Method**

```python
def elbow_method(points):
    # points == data => I use the word of "points" , because it relates to the plot
    
    # distortions => array that containing distortions value
    # distortion => the sum of squared distances from each point to its assigned centroid/center
    distortions = []
    for k in range(1, 11):
        kmeans = KMeans(K=k, max_iteration=100)
        pred_clusters = kmeans.predict(points)
        centroids = kmeans.get_cluster_centroids(kmeans.clusters)
        current_distortion = 0
        for i in range(len(points)):
            current_centroid = centroids[int(pred_clusters[i])]
            current_distortion += ((points[i, 0] - current_centroid[0]) ** 2) + ((points[i, 1] - current_centroid[1]) ** 2)
        distortions.append(current_distortion)
    return distortions
```

```python
# Plot the elbow method
array_k=np.arange(1,11,1)
plt.plot(array_k,elbow_method(X))
plt.xticks(np.arange(1, 11, 1))
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sums of Squares -> WCSS')
plt.title('Elbow method (optimum numbers of clusters)')
plt.show()
```
